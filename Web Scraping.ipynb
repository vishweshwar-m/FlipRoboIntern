2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of
release)

```python
url=['https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc',
'https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt']
name_list=[]
year_list=[]
rating=[]
for u in url:
    page=requests.get(u)
    soup=BeautifulSoup(page.content)
    name=soup.find_all('h3',class_='lister-item-header')
    rate=soup.find_all('div',class_='inline-block ratings-imdb-rating')
    for i in name:
        for j in i.find_all('a'):
            name_list.append(j.text)
        for k in i.find_all('span',class_='lister-item-year text-muted unbold'):
            year_list.append(k.text.replace('(','').replace(')','')) 
    for i in rate:
        for j in i.find_all('strong'):
            rating.append(j.text)
imdb=pd.DataFrame({'Movie Name':name_list,'IMDB Rating':rating,'Year of Release':year_list})
imdb
```

3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year
of release)

```python
page=requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
soup=BeautifulSoup(page.content)
name=soup.find_all('td',class_='titleColumn')
rate=soup.find_all('td',class_='ratingColumn imdbRating')
name_list=[]
year_list=[]
rating=[]
for i in name:
    for j in i.find_all('a'):
        name_list.append(j.text)
    for k in i.find_all('span',class_='secondaryInfo'):
        year_list.append(k.text.replace('(','').replace(')','')) 
for i in rate:
    for j in i.find_all('strong'):
        rating.append(j.text)
imdb_india=pd.DataFrame({'Movie Name':name_list[:100],'IMDB Rating':rating[:100],'Year of Release':year_list[:100]})
imdb_india
```

4. Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
page=requests.get('https://bookpage.com/reviews')
soup=BeautifulSoup(page.content)
name=soup.find_all('h4',class_='italic')
author=soup.find_all('p',class_='sans bold')
review=soup.find_all('div',class_='flex-article-content')
genre=soup.find_all('p',class_='genre-links hidden-phone')
name_list=[]
author_list=[]
review_list=[]
genre_list=[]
for i in name:
    name_list.append(i.text.replace('\n','').replace('★',''))
for i in author:
    author_list.append(i.text.replace('\n',''))  
for i in review:
    for j in i.find_all('p'):
        if len(j.text.replace('\n',''))>50:
            review_list.append(j.text.replace('\n',''))
for i in genre:
    genre_list.append(i.text.replace('\n',''))
book=pd.DataFrame({'Book Name':name_list[:5],'Author Name':author_list[:5],'Genre':genre_list[:5],'Book Review':review_list[:5]})
book    
```

5. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have toscrape:

i) Top 10 ODI teams in men’s cricket along with the records for matches, points andrating.

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup=BeautifulSoup(page.content)
first=soup.find_all('tr',class_='rankings-block__banner')
rest=soup.find_all('tr',class_='table-body')
team_name_list=[i.find('span',class_='u-hide-phablet').text for i in first]
position_list=_list=[i for i in range(1,11)]
dummy=[]
matches_list=[i.find('td',class_='rankings-block__banner--matches').text for i in first]
points_list=[i.find('td',class_='rankings-block__banner--points').text for i in first]
rating_list=[i.find('td',class_='rankings-block__banner--rating u-text-right').text.replace('\n','').strip() for i in first]
for i in rest:
    for j in i.find_all('span',class_='u-hide-phablet'):
        team_name_list.append(j.text)
    for j in i.find_all('td',class_='table-body__cell u-center-text'):
        dummy.append(j.text)
    for j in i.find_all('td',class_='table-body__cell u-text-right rating'):
        rating_list.append(j.text)
matches_list[1:]=dummy[::2]
points_list[1:]=dummy[1::2]
odi_men_team_ranking=pd.DataFrame({'Position':position_list ,'Team Name':team_name_list[:10] ,'Matches':matches_list[:10] ,'Points':points_list[:10] ,'Rating':rating_list[:10]})
odi_men_team_ranking
```

8. Write a python program to extract information about the local weather from the National Weather Service
website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day
extended forecast display for the city. The data should include period, short description, temperature and
description

```python
page=requests.get('https://forecast.weather.gov/MapClick.php?CityName=San+Francisco&state=CA&site=MTR&textField1=37.775&textField2=-122.418#.YQGGbI4zbIU')
soup=BeautifulSoup(page.content)
period=soup.find_all('p',class_='period-name')
des=soup.find_all('p',class_='short-desc')
high=soup.find_all('p',class_='temp temp-high')
low=soup.find_all('p',class_='temp temp-low')
period_list=[]
des_list=[]
high_list=[]
low_list=[]
temp_list=[]
for i in period:
    period_list.append(i.text)
for i in des:
    des_list.append(i.text)
for i in high:
    high_list.append(i.text)
for i in low:
    low_list.append(i.text)
a=len(high_list)
b=len(low_list)
if a>b:
    for i in range(a):
        if i<=b-1:
            temp_list.append(high_list[i][-6::])
            temp_list.append(low_list[i][-6::])
        else:
            temp_list.append(high_list[i][-6::])
else:
    for i in range(b):
        if i<=a-1:
            temp_list.append(low_list[i][-6::])
            temp_list.append(high_list[i][-6::])
        else:
            temp_list.append(low_list[i][-6::])
weather=pd.DataFrame({'Period':period_list,'Description':des_list,'Temperature':temp_list})
weather
```

9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title,
company name, CTC, and apply date

```python
page=requests.get('https://internshala.com/fresher-jobs')
soup=BeautifulSoup(page.content)
tp=soup.find('span',id='total_pages').text
position_name_list=[]
company_name_list=[]
ctc_apply=[]
ctc_name_list=[]
apply_date_list=[]

for i in range(int(tp)):
    page=requests.get('https://internshala.com/fresher-jobs/page-'+str(i))
    soup=BeautifulSoup(page.content)
    pos=soup.find_all('div',class_='heading_4_5 profile')
    comp=soup.find_all('a',class_='link_display_like_text')
    ctc=soup.find_all('div',class_='item_body')
    for i in pos:
        position_name_list.append(i.text.replace('\n','')) 
    for i in comp:
        company_name_list.append(i.text.replace('\n','').strip())
    for i in ctc:
        ctc_apply.append(i.text.replace('\n','').strip())
apply_date_list=ctc_apply[2::3]
ctc_name_list=ctc_apply[1::3]
job=pd.DataFrame({'Position':position_name_list,'Company Name':company_name_list,'CTC':ctc_name_list,'Apply Date':apply_date_list})
job.drop_duplicates(inplace=True)
job.reset_index(inplace=True,drop=True)
job
```
